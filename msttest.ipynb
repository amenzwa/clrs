# Tests of Minimum Spanning Tree Algorithms

This notebook contains tests for the algorithms implemented in the `mst.ipynb` notebook.

## test Kruskal's MST

Let us test `mstKruskal()` against the connected, undirected, weighted graph shown in Figure 21.4 p.593. First we construct the graph.

```python
from graph import *
from mst import *

def makeKG() -> MSTGraph:
    vs = ["a", "b", "c", "d", "e", "f", "g", "h", "i"]
    es = {
        "a": ["b", "h"],
        "b": ["a", "c", "h"],
        "c": ["b", "d", "f", "i"],
        "d": ["c", "e", "f"],
        "e": ["d", "f"],
        "f": ["c", "d", "e", "g"],
        "g": ["f", "h", "i"],
        "h": ["a", "b", "g", "i"],
        "i": ["c", "g", "h"],
    }
    ew = {
        "a-b": 4,
        "a-h": 8,
        "b-a": 4,
        "b-c": 8,
        "b-h": 11,
        "c-b": 8,
        "c-d": 7,
        "c-f": 4,
        "c-i": 2,
        "d-c": 7,
        "d-e": 9,
        "d-f": 14,
        "e-d": 9,
        "e-f": 10,
        "f-c": 4,
        "f-d": 14,
        "f-e": 10,
        "f-g": 2,
        "g-f": 2,
        "g-h": 1,
        "g-i": 6,
        "h-a": 8,
        "h-b": 11,
        "h-g": 1,
        "h-i": 7,
        "i-c": 2,
        "i-g": 6,
        "i-h": 7,
    }
    g = MSTGraph("Kruskal")
    g.makeVE(vs, es, ew)
    return g

kg = makeKG()
print(kg)
draw(kg, directed=False, label=f"{kg.id} connected, undirected graph")
```

Then, we apply `mstKruskal()` to the graph.

```python
kt = mstKruskal(kg)
print(kt)
draw(kt, directed=False, label=f"{kt.id} Minimum Spanning Tree")
```
